```python
import pandas as pd 
```

```python
dataPonsel = pd.read_csv('dataPonsel.csv')
```

```python
data = dataPonsel.copy()
```

```python
target = 'kecanduan'
```

```python
encode = ['gender', 'jenjangPekerjaan','intensitasSosmed','gangguanTidur','cemasPonselTertinggal','gangguanProduktivitas']
```

```python
data
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data,dummy], axis=1)
    del data[col]
```

```python
data
```

```python
target_mapper = {'Ya':0, 'Tidak':1}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['kecanduan'] = data['kecanduan'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('kecanduan', axis=1)
Y = data['kecanduan']
```

```python
X
```

```python
Y
```

```python
#Model NBC 
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, Y)
```

```python
# menyimpan moden nbc
import pickle 
pickle.dump(model, open('modelNBC_dataPonsel.pkl','wb'))
```
